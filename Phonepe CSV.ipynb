```python
import os

path = r"D:\My files\Nagammai\Guvi\MDE86\code\clone\data\aggregated\transaction\country\india\state"
Ag_state_list = os.listdir(path)

column = {'State': [], 'Year': [], 'Quarter': [], 'Transaction_type': [], 'Transaction_count': [], 'Transaction_amount': []}

for state in Ag_state_list:
    state_path = os.path.join(path, state)
    Ag_yr = os.listdir(state_path)
    
    for year in Ag_yr:
        year_path = os.path.join(state_path, year)
        Ag_yr_list = os.listdir(year_path)
        
        for quarter in Ag_yr_list:
            quarter_path = os.path.join(year_path, quarter)
            with open(quarter_path, 'r') as file:
                data = json.load(file)
                for item in data['data']['transactionData']:
                    name = item['name']
                    count = item['paymentInstruments'][0]['count']
                    amount = item['paymentInstruments'][0]['amount']
                    
                    column['Transaction_type'].append(name)
                    column['Transaction_count'].append(count)
                    column['Transaction_amount'].append(amount)
                    column['State'].append(state)
                    column['Year'].append(year)
                    column['Quarter'].append(int(quarter.strip('.json')))

# Successfully created a dataframe
Ag_Trans = pd.DataFrame(column)
Ag_Trans.to_csv('aggregated_transaction.csv', index=False)
```

```python
import os
import json
import pandas as pd

path6 = r"D:\My files\Nagammai\Guvi\MDE86\code\clone\data\top\user\country\india\state"
USER_list = os.listdir(path6)

col6 = {'State': [], 'Year': [], 'Quarter': [], 'District': [],
        'RegisteredUser': []}
for i in USER_list:
    p_i = os.path.join(path6, i) + "/"
    Agg_yr = os.listdir(p_i)

    for j in Agg_yr:
        p_j = os.path.join(p_i, j) + "/"
        Agg_yr_list = os.listdir(p_j)

        for k in Agg_yr_list:
            p_k = os.path.join(p_j, k)
            # Extract year and quarter from file path
            year = int(j)
            quarter = int(k.strip('.json'))
            # Load JSON data
            with open(p_k, 'r') as file:
                data = json.load(file)
                # Extract district information from pincodes list
                for district_data in data['data']['pincodes']:
                    district_name = district_data['name']
                    registered_users = district_data['registeredUsers']
                    # Append data to column dictionary
                    col6['State'].append(i)
                    col6['Year'].append(year)
                    col6['Quarter'].append(quarter)
                    col6['District'].append(district_name)
                    col6['RegisteredUser'].append(registered_users)

df_top_user = pd.DataFrame(col6)
df_top_user.to_csv('top_user.csv', index=False)
```
